```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# NumPy Operations
import numpy as np # Added import statement here


# 1. Creating Arrays
print(" NumPy: Creating Arrays")
array_1d = np.array([1, 2, 3, 4, 5])  # 1D array
array_2d = np.array([[1, 2, 3], [4, 5, 6]])  # 2D array
zeros = np.zeros((2, 3))  # 2x3 matrix of zeros
ones = np.ones((3, 3))  # 3x3 matrix of ones

print("1D Array:", array_1d)
print("2D Array:\n", array_2d)
print("Zeros Array:\n", zeros)
print("Ones Array:\n", ones)

# 2. Mathematical Operations
print("\nNumPy: Mathematical Operations")
array = np.array([1, 2, 3, 4])
print("Original Array:", array)
print("Array + 5:", array + 5)
print("Array * 2:", array * 2)
print("Array squared:", array ** 2)

# 3. Aggregate Functions
print("\nNumPy: Aggregate Functions")
print("Sum:", np.sum(array))
print("Mean:", np.mean(array))
print("Standard Deviation:", np.std(array))

# 4. Reshaping and Transposing
print("\nNumPy: Reshaping and Transposing")
reshaped = array_2d.reshape((3, 2))
print("Reshaped Array:\n", reshaped)
print("Transposed Array:\n", array_2d.T)

# 5. Random Numbers
print("\n NumPy: Random Numbers")
random_array = np.random.rand(3, 3)  # Random values in [0, 1]
print("Random Array:\n", random_array)

```

```python
#Pandas Operations


# 1. Creating a DataFrame
print("\nPandas: Creating a DataFrame")
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 29],
    'Score': [85, 90, np.nan, 88]
}
df = pd.DataFrame(data)
print("DataFrame:\n", df)

# 2. Handling Missing Data
print("\n Pandas: Handling Missing Data")
print("Original DataFrame:\n", df)
df['Score'].fillna(df['Score'].mean(), inplace=True)  # Fill missing values with mean
print("After Filling Missing Data:\n", df)

# 3. Filtering Rows
print("\nPandas: Filtering Rows")
filtered_df = df[df['Age'] > 25]  # Filter rows where Age > 25
print("Filtered DataFrame:\n", filtered_df)

# 4. Grouping and Aggregation
print("\n Pandas: Grouping and Aggregation")
grouped_df = df.groupby('Age')['Score'].mean()  # Group by Age and calculate mean Score
print("Grouped DataFrame:\n", grouped_df)

# 5. Sorting Data
print("\nPandas: Sorting Data")
sorted_df = df.sort_values(by='Score', ascending=False)
print("Sorted DataFrame:\n", sorted_df)

# 6. Reading and Writing CSV
print("\nPandas: Reading and Writing CSV")
df.to_csv('output.csv', index=False)  # Save to CSV
loaded_df = pd.read_csv('output.csv')  # Read from CSV
print("Loaded DataFrame:\n", loaded_df)




```
